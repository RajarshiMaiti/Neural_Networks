Preprocessing

Inferences:

To begin preprocessing we will create a function called preprocess_inputs that will make a copy of a dataframe. month and day are categorical features that take an ordering.

We updated preprocess_inputs by creating another function and calling ordinal_code to ordinal code the month and day columns.

We then will split and scale the data. Because we are treating this as a regression and classification task, we include a task= variable in the preprocess_inputs function by defining which task we want to run.

```python
def ordinal_code(df, column, ordering):
    df = df.copy()
    df[column] = df[column].apply(lambda x: ordering.index(x))
    return df
```

```python
def preprocess_inputs(df, task='classification'):
    df = df.copy()
    
    # ordinal encode month and day columns
    df = ordinal_code(
        df,
        column='month',
        ordering=[
            'jan',
            'feb',
            'mar',
            'apr',
            'may',
            'jun',
            'jul',
            'aug',
            'sep',
            'oct',
            'nov',
            'dec'
        ]
    )
    df = ordinal_code(
        df,
        column='day',
        ordering=[
            'sun',
            'mon',
            'tue',
            'wed',
            'thu',
            'fri',
            'sat'
        ]
    )
    
    # split df into X and y
    if task == 'regression':
        y = df['area']
    elif task == 'classification':
        y = df['area'].apply(lambda x: 1 if x > 0 else 0) # will create a new col that will suggest if it is a large fire or small fire
    
    X = df.drop('area', axis=1) # all other data
    
    # train test split
    X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.7, shuffle=True, random_state=1)
    
    # scaler feature data (X)
    scaler = StandardScaler()
    scaler.fit(X_train)
    
    X_train = pd.DataFrame(scaler.transform(X_train), columns=X.columns)
    X_test = pd.DataFrame(scaler.transform(X_test), columns=X.columns)
        
    return X_train, X_test, y_train, y_test
```

Classification

```python
X_train, X_test, y_train, y_test = preprocess_inputs(data, task='classification')
```

We can see that running a basic Linear Regression Model, our R^2 is 0.02 - which is not really good either. We can create a complex model using Neural Network.



```python
lin_reg_model = LinearRegression()
lin_reg_model.fit(X_train, y_train)

print("Linear Regression R^2: {:.5f}".format(lin_reg_model.score(X_test, y_test)))
```

When first running the neural network model, we know that we must train a neural network with the data in order for the model to improve. But in the first instances of training, the R^2 is not improving. That is because the neural network model works best if data is scaled and standardize the data. So to standardize the data we included it in our preprocess_inputs function using the StandardScaler from sklearn.



```python
nn_reg_model = MLPRegressor(hidden_layer_sizes=(16,16))
nn_reg_model.fit(X_train, y_train)

print("NN Regression R^2: {:.5f}".format(nn_reg_model.score(X_test, y_test)))
```
